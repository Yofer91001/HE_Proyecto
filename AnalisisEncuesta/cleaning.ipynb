```python
import pandas as pd
```

```python
df = pd.read_excel('data/raw/raw_data.xlsx')
```

```python
nuevas_columnas = ['actividad','Ventas','ventas_post','personas','personas_post','locales','residenciales','tiempo','tiempo_post'] 
for i in range(9):
    df = df.rename(columns={
        df.columns[i+1]:nuevas_columnas[i] 
    })
df.sort_values(by='actividad', ascending=False) 
```

```python
for c in df.columns[2:]:
    for v in df[c]:
        if v < 0:
            df[c] = df[c]*-1
            break            
```

```python
df['diff_tiempo'] = df['tiempo'] - df['tiempo_post']
```
